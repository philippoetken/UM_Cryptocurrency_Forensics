# Read in one website with multible pages

```python
# @phillip you could built a wrapper here to load the urls from a list and to run it for all of them
# to specify:
#@Phillip you have to adjust this 
dir_save = 'Huobi'
print(dir_save)
#@Phillip you have to adjust this 
start_url = 'https://www.walletexplorer.com/wallet/Huobi.com'


# make output directory
try:
    os.mkdir('output')
except:
    pass

# create subdirectory for this wallet
try:
    os.mkdir('output/{}'.format(dir_save), )
except:
    pass

# set current page to 0 or load from json file
current_page = None
try:
    with open('output/{}/current_page_transaction.json'.format(dir_save)) as json_file:
        current_page = json.load(json_file)
        print("start_with_current_page:",current_page)
except:
    current_page = 0
    print("start_with_current_page:",current_page)	



# get how many pages to scrape
result = requests.get(start_url)
doc = BeautifulSoup(result.text, "html.parser")

page_tag = doc.find(text=re.compile("Page 1 /*") ) #can add more attributes
pages = int(page_tag.split('/')[1].strip())

# loop over all pages and save addresses
scraped_transactions = []
# loop
for i in range(pages):
    #  get scraped addresses to continue scraping
    if i < current_page:
        print('pass page', i)
        with open('output/{}/scraped_transactions.json'.format(dir_save)) as f1:
            scraped_transactions = json.load(f1)
            print('l1',len(scraped_transactions))

        continue
    
    
    print(i,'/',pages)
    if i == 0:
        url = start_url
    else:
        url = start_url + '?page='+str(i+1)

    result = requests.get(url)

    doc = BeautifulSoup(result.text, "html.parser")

    transactions = doc.find_all(class_="txid")

     # stop if no addresses found
    if len(transactions) == 0:
        print('no addresses found in:', url)
        break
    else:
        print('found', len(transactions), 'transactions: ', url)

    for transaction in transactions:
        txid = transaction.find('a').get('href').split('/')[-1]
        scraped_transactions.append(txid)


    # print output to file
    with open('output/{}/scraped_transactions.json'.format(dir_save), 'w') as f:
        json.dump(scraped_transactions, f)

    # save current page
    current_page = current_page+1
    with open('output/{}/current_page_transaction.json'.format(dir_save), 'w') as outfile:
        json.dump(current_page, outfile)
    
    # adjust this to specify minutes to sleep
    #sleep(randint(1,5))
```
