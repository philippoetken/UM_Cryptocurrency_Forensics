## scrape all wallets of interests

```python
def scrape(category, inputDf):
    p = 0
    for index, row in inputDf.iterrows():
        folder = category + "/" + str(p) + '_' + inputDf[category][index]
        p = p+1

        dir_save = folder
        print(dir_save)

        start_url = inputDf['we-link'][index]

        # make output directory
        try:
            os.mkdir('output/' + category)
        except:
            pass
                
        # create subdirectory for this wallet
        try:
            os.mkdir('output/{}'.format(dir_save), )
        except:
            pass
                
        # set current page to 0 or load from json file
        current_page = None
        try:
            with open('output/{}/current_page.json'.format(dir_save)) as json_file:
                current_page = json.load(json_file)
                print("start_with_current_page:",current_page)
        except:
            current_page = 0
            print("start_with_current_page:",current_page)	

        # get how many pages to scrape
        try:
            result = requests.get(start_url)
            doc = BeautifulSoup(result.text, "html.parser")
            page_tag = doc.find(text=re.compile("Page 1 /*")) #can add more attributes
            pages = int(page_tag.split('/')[1].strip())
        except:
            errorWait = randint(120,180)
            print('Disconnected on getting page amount. Continue in ' + str(errorWait) + ' seconds.')
            sleep(errorWait)
            
            result = requests.get(start_url)
            doc = BeautifulSoup(result.text, "html.parser")
            page_tag = doc.find(text=re.compile("Page 1 /*")) #can add more attributes
            pages = int(page_tag.split('/')[1].strip())         

        # loop over all pages and save addresses
        scraped_addresses = []

        # set sleeptimer to avoid disconnection by walletexplorer
        sleeptimerReset = 95
        sleeptimer = sleeptimerReset

        # loop
        for i in range(pages):
                    
            #  get scraped addresses to continue scraping
            if i < current_page:
                print('pass page', i)
                continue
                    
            try:
                with open('output/{}/scraped_addresses.json'.format(dir_save)) as f1:
                    scraped_addresses = json.load(f1)
            except:
                pass

            # make backup every 100 pages
            if i%100 == 0:
                try:
                    with open('output/{}/scraped_addresses.json'.format(dir_save)) as fb:
                        scraped_addresses_backup = json.load(fb)
                    with open('output/{}/scraped_addresses_backup.json'.format(dir_save), 'w') as fbb:
                        scraped_addresses_backup.extend('{Last page: '+str(i)+'}')
                        json.dump(scraped_addresses_backup, fbb)        
                except:
                    pass

            sleeptimer = sleeptimer-1
                    
            print(i,'/',pages)
                    
            if i == 0:
                url = start_url
            else:
                url = start_url + '?page='+str(i+1)
                    
            try: 
                result = requests.get(url)
                doc = BeautifulSoup(result.text, "html.parser")
                
                # get addresses with regex
                addresses = doc.find_all(text=re.compile("^(bc1|[13])[a-zA-HJ-NP-Z0-9]{25,39}$") )
                print('found', len(addresses), 'addresses in:', url)
                scraped_addresses.extend(addresses)
                sleep(1)

                if len(addresses) == 0 and i < (pages-1):
                    raise Exception('')
            except:
                errorWait = randint(120,180)
                print('Disconnected on getting addresses. Continue in ' + str(errorWait) + ' seconds.')
                sleep(errorWait)
                
                result = requests.get(url)
                doc = BeautifulSoup(result.text, "html.parser")

                # get addresses with regex
                addresses = doc.find_all(text=re.compile("^(bc1|[13])[a-zA-HJ-NP-Z0-9]{25,39}$") )
                print('found', len(addresses), 'addresses in:', url)
                scraped_addresses.extend(addresses)
                sleep(1)

            # stop if no addresses found
            if len(addresses) == 0:
                print('no addresses found in:', url)
                break
                    
            # print output to file
            with open('output/{}/scraped_addresses.json'.format(dir_save), 'w') as f:
                json.dump(scraped_addresses, f)
                sleep(1)

            # save current page
            current_page = current_page+1
            with open('output/{}/current_page.json'.format(dir_save), 'w') as outfile:
                json.dump(current_page, outfile)

            # adjust this to specify seconds to sleep
            if sleeptimer > 0:
                print('remaining pages until break: '+ str(sleeptimer))
                sleep(randint(1,3))
            else:
                print('remaining pages until break: '+ str(sleeptimer))
                sleeptimer = sleeptimerReset
                pauseWait=randint(60,180)
                print('Waiting. Continue in ' + str(pauseWait) + ' seconds.')
                sleep(pauseWait)
```
